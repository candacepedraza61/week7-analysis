```python
## Improving memory allocation
# if we were working with a massive dataset, we'd want to find ways to improve processing power
# for example, the following columns have a few values that repeat again and again as strings. this is highly inefficient

#season column has winter and summer
#medal column has gold, silver and bronze
#gender at this point in the olympics only has male and female

#one way to improve memory allocation is to take columns with that contain only a few data points and turn them into categories
```

```python
## you should see how season, medal and gender are all string objects
#note that memory usage: 3.1+MB

oly.info()
```

```python
#convert gender, medal and season to category
#changing the type of python object it is

oly[["Gender","Medal","Season"]] = oly[["Gender","Medal","Season"]].astype("category") 
```

```python
#see what the memory allocation is now using info()

oly.info()
```

```python
#Filtering by dates
#You'll learn more about date and time in the coming week. For now, please note how the Year column is a int64 object
#and NOT a datetime object. We can still work with it for our needs.
```

```python
#Find all the 1900 competitions
oly[oly["Year"] == 1900] 
```

```python
## return all competitions between 1920 and 1950 (inclusive)

##METHOD 1
oly[(oly["Year"] >= 1920) & (oly["Year"] <= 1950)]
```

```python
# return the 1920 through 1950 competitions

#METHOD 2

#Declare your filters

my_filter_1 = oly["Year"] >= 1920
my_filter_2 = oly["Year"] <= 1950

##Filter

oly[my_filter_1 & my_filter_2]
```

```python
## return only Tennis competitions between 1920 and 1950 (inclusive) 
#METHOD 2

#declare your filters

my_filter_1 = oly["Year"] >= 1920
my_filter_2 = oly["Year"] <= 1950
my_filter_3 = oly["Sport"] == "Tennis"

## Filter

oly[my_filter_1 & my_filter_2 & my_filter_3]
```

```python
#method 3

#declare filters with meaningful names
f_yr1920 = oly["Year"] >= 1920
f_yr1950 = oly["Year"] <= 1950
f_tennis = oly["Sport"] == "Tennis"

#Filter

oly[f_yr1920 & f_yr1950 & f_tennis]
```

```python
##return just women's tennis competitions between 1920 and 1950

f_yr1920 = oly["Year"] >= 1920
f_yr1950 = oly["Year"] <= 1950
f_tennis = oly["Sport"] == "Tennis"
f_women = oly["Gender"] == "Women"

oly_tennis = oly[f_yr1920 & f_yr1950 & f_tennis & f_women]
oly_tennis.tail()
```
